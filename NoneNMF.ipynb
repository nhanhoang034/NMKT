```python
import pandas as pd
```

```python
#Đọc dữ liệu
Minerals_Database = pd.read_csv('Minerals_Database.csv')

#Lọc các cột cần dùng và các hàng có cả 3 giá trị Magnesium, Calcium, Iron bằng 0
Minerals_Database_filter = Minerals_Database[['Name', 'Magnesium', 'Calcium', 'Iron', 'Molar Mass']]
Minerals_Database_use = Minerals_Database_filter[(Minerals_Database_filter['Magnesium'] > 0) 
                                                | (Minerals_Database_filter['Calcium'] > 0) 
                                                | (Minerals_Database_filter['Iron'] > 0)]
display(Minerals_Database_use)
```

```python
Minerals_Database_sort['MgO_Percent'] = (Minerals_Database_sort['Magnesium']*40 / Minerals_Database_sort['Molar Mass']) * 100
Minerals_Database_sort['CaO_Percent'] = (Minerals_Database_sort['Calcium']*56  / Minerals_Database_sort['Molar Mass']) * 100
Minerals_Database_sort['FeO_Percent'] = (Minerals_Database_sort['Iron']*72 / Minerals_Database_sort['Molar Mass']) * 100

# Hiển thị kết quả với phần trăm của Mg, Ca, Fe trong Molar Mass
top_minerals_percent = Minerals_Database_sort[['Name', 'MgO_Percent', 'CaO_Percent', 'FeO_Percent']].head(10)
display(top_minerals_percent)
```

```python
# Tính phần trăm CO2 từ các oxit
Minerals_Database_sort['CO2'] = ( Minerals_Database_sort['MgO_Percent'] * 1.092 
                                + Minerals_Database_sort['CaO_Percent'] * 0.785 
                                + Minerals_Database_sort['FeO_Percent'] * 0.859)

# Hiển thị kết quả với phần trăm CO2
top_minerals_CO2 = Minerals_Database_sort[['Name','MgO_Percent','CaO_Percent','FeO_Percent', 'CO2']].head(10)
display(top_minerals_CO2)
```
