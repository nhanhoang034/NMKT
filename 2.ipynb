```python
import pandas as pd
```

```python
#Đọc dữ liệu
Minerals_Database = pd.read_csv('Minerals_Database.csv')

#Lọc các cột cần dùng và các hàng có cả 3 giá trị Magnesium, Calcium, Iron bằng 0
Minerals_Database_filter = Minerals_Database[['Name', 'Magnesium', 'Calcium', 'Iron', 'Molar Mass']]
Minerals_Database_use = Minerals_Database_filter[(Minerals_Database_filter['Magnesium'] > 0) 
                                                | (Minerals_Database_filter['Calcium'] > 0) 
                                                | (Minerals_Database_filter['Iron'] > 0)]
Minerals_Database_sort = Minerals_Database_use.sort_values(by=['Magnesium'], ascending=False)

top_5_minerals = Minerals_Database_sort.head(5)
# Hiển thị kết quả sau khi lọc
display(top_5_minerals)
```

```python
# Tính phần trăm của Magnesium, Calcium và Iron trong Molar Mass
Minerals_Database_sort['Mg_Percentage'] = (Minerals_Database_sort['Magnesium']*24 / Minerals_Database_sort['Molar Mass']) * 100
Minerals_Database_sort['Ca_Percentage'] = (Minerals_Database_sort['Calcium']*40  / Minerals_Database_sort['Molar Mass']) * 100
Minerals_Database_sort['Fe_Percentage'] = (Minerals_Database_sort['Iron']* 56 / Minerals_Database_sort['Molar Mass']) * 100

# Hiển thị kết quả với phần trăm của Mg, Ca, Fe trong Molar Mass
top_5_minerals_percentage = Minerals_Database_sort[['Name', 'Mg_Percentage', 'Ca_Percentage', 'Fe_Percentage']].head(5)
display(top_5_minerals_percentage)
```

```python
# Tính phần trăm MgO, CaO và FeO
Minerals_Database_sort['MgO_Percentage'] = Minerals_Database_sort['Mg_Percentage'] * 1.092
Minerals_Database_sort['CaO_Percentage'] = Minerals_Database_sort['Ca_Percentage'] * 0.785
Minerals_Database_sort['FeO_Percentage'] = Minerals_Database_sort['Fe_Percentage'] * 0.859

# Tính phần trăm CO2 từ các oxit
Minerals_Database_sort['CO2'] = (Minerals_Database_sort['MgO_Percentage'] 
                                              + Minerals_Database_sort['CaO_Percentage'] 
                                              + Minerals_Database_sort['FeO_Percentage'])

# Hiển thị kết quả với phần trăm CO2
top_5_minerals_CO2 = Minerals_Database_sort[['Name', 'CO2']].head(5)
display(top_5_minerals_CO2)
```
